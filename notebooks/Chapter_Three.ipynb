# Hypothesis Testing

There is a common belief one should always roll with a score of less than 350 points, regardless of the number of dice in play. Likewise, I have always felt that I should stop rolling whenever my score is above 1000. Is this the best playstyle? Let's find out!

Run option 3 on station.py. It will generate mean scores for our vps.  You will see csv files in the data directory which contain the mean scores for each vp. I've included only three: Perfect, Random, and myself (Gavin). On the next cell, we will plot the data and see how they compare.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# add any additional VPs that you make to this list
vp_names = ['perfect', 'random', 'gavin']

fig, ax = plt.subplots()
for vp in vp_names:
    df = pd.read_csv(f'../data/{vp}.csv')
    ax = sns.distplot(df, label=vp)
    
ax.legend()
ax.set_ylabel('Scaled as likelihood', fontsize=16)
ax.set_xlabel('Average mean score per test', fontsize=16)
fig.suptitle('Comparison of VPs by mean test scores', fontsize=20)
plt.savefig('../images/vp_compare.png', format='png')
plt.show()
plt.close()
```

Finally, we get to start comparing the VPs! And the results couldn't be more clear. I am definitely better than the Random VP, which randomly chooses to roll or stay. But I'm still a long way off from Perfect. This is probably what you expected.

*Remember, the Perfect VP cheats, and is not actually attainable. We are only including it here to give us a clear upper boundary on our graph.*
