```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import seaborn as sns
from scipy.stats import norm, ks_2samp
from statsmodels.stats.power import tt_ind_solve_power as power
```

```python
# This helper function creates Matplotlib figures for the
# histograms used in this document.
def hist_helper(ax, x_label):
    ax.set_ylim(0,.0025)
    ax.set_xlim(0,1500)
    ax.set_xlabel(x_label, fontsize=12)
    ax.set_ylabel('Density Of Scores', fontsize=12) 
```

```python
# This function reduces scores with the outcome False: 
# When a player loses the game, there score must be 0.
# Also turns the pandas df into a np array
def reduce(file_name):
    df = pd.read_csv(f'../data/{file_name}.csv')
    df.score.where(df.outcome,0,inplace=True)
    return np.array(df.score)
```

```python
# This cell creates a histogram showing the
# distribution of points from 5 million rounds
# of boxcar using the Perfect VP.
fig, ax = plt.subplots()
df = pd.read_csv('../data/boxcar.csv')
ax = sns.distplot(df['score'])
ax = hist_helper(ax, 'Points From One Round of Boxcar') 
plt.suptitle('Distribution of scores from 5,000,000 rounds', y=1 )
plt.savefig('../images/sim_size_test.png', format='png')
plt.tight_layout()
plt.show()
plt.close()
```

```python
# This cell creates a scatter plot showing how
# big the simulation sample size should be.

# Reading in data and calculating mean / var
fig, ax = plt.subplots()
df = pd.read_csv('../data/batchtestold.csv')
mean = df.groupby('batch_size').mean()
print(mean)
var = df.groupby('batch_size').var()

# Plotting data
ax = sns.scatterplot(df['batch_size'], 
                     df['mean'], 
                     alpha=.5,
                     label='Test mean')
ax = sns.scatterplot(mean['mean'].index,
                     mean['mean'],
                     color = 'r',
                     label='Avg test mean')
ax = sns.scatterplot(var['mean'].index,
                     var['mean'],
                     color='b',
                     label='Variance of tests')

# Making the plot look better
ax.set_xlabel('Number of scores per test', fontsize=12)
ax.set_ylabel('Variance / Average score', fontsize=12) 
plt.ylim(bottom=0, top=800);
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
fig.suptitle('How many simulations should be run?', fontsize=20)
ax.axvline(8000, color='k', linestyle=':')
ax.text(8000,150, '<---8k samples needed')
plt.legend()
plt.show()
plt.savefig('../images/sim_size_test.png', format='png')
plt.close()
```

```python
# This cell is similar to the one above, but focuses on 
# number of bootstraps needed

# Loading in data and determining mean / var
scores = pd.read_csv('../data/boottests.csv')
grp_mean = scores.groupby('size')['mean'].mean()
grp_var = scores.groupby('size')['mean'].var()

# Plotting data
fig, ax = plt.subplots()
ax = sns.scatterplot(scores['size'],
                     scores['mean'],
                     alpha=.5,
                     label='Test mean')
ax = sns.scatterplot(grp_mean.index,
                     grp_mean,
                     color ='r',
                     label='Avg test mean')
ax = sns.scatterplot(grp_var.index,
                     grp_var,
                     color ='b',
                     label='Variance of tests')

# Improving look of the graph
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
plt.ylim(bottom=0, top=800);
plt.legend()
ax.set_ylabel('Variance / Average Score', fontsize=12)
ax.set_xlabel('Number of bootstraps taken', fontsize=12)
fig.suptitle('How many bootstraps per dataset?', fontsize=20)
ax.axvline(8000, color='k', linestyle=':')
ax.text(8000,150, '<-8k bootstraps needed')
plt.savefig('../images/boot_size_test.png', format='png')
;
```

```python
# This helper function takes an array of data, performs a bootstrap, 
# And returns a scipy.stats.norm distribution object based on it.
def boot_dist(array):
    mean = []
    std = []
    for _ in range(8000):
        boot = np.random.choice(a=array,size=8000,replace=True)
        mean.append(boot.mean())
        std.append(boot.std())
    return norm(np.mean(mean), (np.mean(std)/np.sqrt(8000)) )
```

```python
# This helper function takes an axes, a tuple of information, and 
# a range of x values. It utilizes boot_dist to create a dist object,
# plots the distribution on the axes and adds labels. The dist is returned.
def dist_plot(ax, tup, x):
    if tup[3]:
        df = reduce(tup[0])
    else:
        df = np.array(pd.read_csv(tup[0]).score)
    dist = boot_dist(df)
    ax.plot(x,dist.pdf(x),linewidth=2,label=tup[1])
    ax.axvline(df.mean(),color=tup[2],linestyle=':')
    return dist
```

```python
# Here we are plotting the the distributions of means for 
# each of our VPs.
fig, ax = plt.subplots()
x = np.linspace(0, 2000, num=2000)
for tup in [('random','VP: Random','C0', True),
            ('wyatte', 'VP: Wyatte', 'C1', True),
            ('karen', 'VP: Karen', 'C2', True),
            ('judy', 'VP: Judy', 'C3', True),
            ('../data/perfect.csv', 'VP: Perfect', 'C4', False)]:
    dist_plot(ax, tup, x)
ax.set_xlim(300,700)
ax.legend()
ax.set_ylabel('Probability of scoring', fontsize=16)
ax.set_xlabel('Average points per round', fontsize=16)
fig.suptitle('Distributions of VP Means', fontsize=20)
plt.savefig('../images/vp_compare.png', format='png')
plt.show()
plt.close()
```

```python
# Here we are plotting the distribution of Means for only
# three data sets.
fig, ax = plt.subplots()
x = np.linspace(0, 2000, num=2000)
for tup in [('karen', 'VP: Karen', 'C0', True),
            ('judy', 'VP: Judy', 'C1', True),
            ('GavinRay', 'HP: Gavin Ray', 'C2', True)]:
    dist_plot(ax, tup, x)
ax.set_xlim(350,550)
ax.legend()
ax.set_ylabel('Probability of scoring', fontsize=16)
ax.set_xlabel('Average points per round', fontsize=16)
fig.suptitle('Gavin Ray Score Means Comparison', fontsize=20)
plt.savefig('../images/hp_compare.png', format='png')
plt.show()
plt.close()
```

```python
# This cell plots the actual scores as histograms for Gavin's first
# set of scores, and the Judy VP scores.

# Bringing in data with reduce() helper function
null =reduce('GavinRay')
alt = reduce('judy')

# Plotting scores from GavinRay.csv and judy.csv as histograms.
fig, axes = plt.subplots(1,2)
hist_helper(axes[0], 'Gavin\'s Scores')
hist_helper(axes[1], 'Judy VP Scores')
sns.distplot(null, ax=axes[0])
sns.distplot(alt, ax=axes[1])

# Adding shading and dotted lines to areas of interest
axes[0].fill_between([100, 300], [.003,.003], color='y', alpha=.15)
axes[0].fill_between([900, 1100], [.003,.003], color='y', alpha=.15)
axes[1].fill_between([100, 300], [.003,.003], color='y', alpha=.15)
axes[1].fill_between([900, 1100], [.003,.003], color='y', alpha=.15)
axes[1].set_yticklabels([])
for each in [100,300,900,1100]:
    axes[0].axvline(each, color='k', linestyle=':')
    axes[1].axvline(each, color='k', linestyle=':')
    
# Finishing the graphs
fig.suptitle('Showing differences Between Gavin And Judy', y=1)
fig.tight_layout()
plt.savefig('../images/hp_jd_compare.png', format='png')
plt.show()
plt.close()
```

These two graphs show the differences between our human player's scores and the Judy VP. Mainly, our player is making more scores in the 100 to 300 point range as indicated by the yellow bar. Judy, on the other hand, is giving up these scores. As a result, Judy is able to gain substantially more rolls with scores between 900 and  1100.

We will now give our human player a second try after informing them of these new revelations.

```python
# This cell makes a similar histogram to the ones above, but only
# For Gavin's second set of scores.

# Bring in data with reduce()
h2 = reduce('GavinRay2')

# Plotting the graphs
fig, ax = plt.subplots()
hist_helper(ax, 'Scores from My Second Attempt')
sns.distplot(h2, ax=ax)

# Adding flair to areas of interest
ax.fill_between([100, 300], [.003,.003], color='y', alpha=.15)
ax.fill_between([900, 1100], [.003,.003], color='y', alpha=.15)
for each in [100,300,900,1100]:
    ax.axvline(each, color='k', linestyle=':')
    
# Finishing up.
fig.suptitle('An Improvement in Scores.', y=1)
fig.tight_layout()
plt.savefig('../images/hp2_compare.png', format='png')
plt.show()
plt.close()
```

```python
# Here I am doing a hypothesis test. The null distribution is Gavin's first
# set of scores. The alt distribution is from Judy VP. Both are being 
# compared to Gavin's second set of scores, referred to as sample.
null_vs_sample_p_score = ks_2samp(h2, null)
alt_vs_sample_p_score = ks_2samp(h2, np.random.choice(alt,500))
print(null_vs_sample_p_score)
print(alt_vs_sample_p_score)
```

```python
# This cell compares the distribution of means from Gavin's first set of scores, 
# the Judy VP, and Gavin's second set of scores. 

# Create graph and plot data
fig, ax = plt.subplots()
x = np.linspace(0, 2000, num=2000)
g_dist = dist_plot(ax, ('GavinRay', 'Old Scores', 'C0', True),x)
j_dist = dist_plot(ax, ('judy', 'VP: Judy', 'C1', True) ,x)
g2_dist = dist_plot(ax, ('GavinRay2', 'New Scores', 'C2', True) ,x)

# Add labels and set dimensions.
ax.set_xlim(400,550)
ax.legend()
ax.set_ylabel('Probability of scoring', fontsize=16)
ax.set_xlabel('Average points per round', fontsize=16)
fig.suptitle('New Score Means Comparison', fontsize=20)


# Shading area shared by two dists
mid_point = (g2_dist.mean() + j_dist.mean())/2
x1 = np.linspace(g2_dist.ppf(.01),mid_point,500)
ax.fill_between(x1, g2_dist.pdf(x1), color='grey', alpha = .25)
x2 = np.linspace(mid_point,j_dist.ppf(.99), 500)
ax.fill_between(x2, j_dist.pdf(x2),color='grey', alpha = .25)

# Determining size of shared region.
p_of_both = round((g2_dist.cdf(mid_point) ) + (1-j_dist.cdf(mid_point)),2)
ax.text(485, .03, f'Joint Probabilities:\n      {p_of_both}')

# Save and close figure
plt.savefig('../images/hp3_compare.png', format='png')
plt.show()
plt.close()

```
